```python
import numpy as np
import pandas as pd
```

```python
# Load the data
data = pd.read_csv('../data/spectral_gap.csv')
```

```python
data.columns
```

```python
n = data[' n']
spectral_gap = data[' spectral_gap']
```

```python
# plot n and spectral_gap
import matplotlib.pyplot as plt
plt.plot(n, spectral_gap)
plt.xlabel('n')
plt.ylabel('spectral_gap')
```

```python
# fit n and spectral_gap using exponential (scipy)
from scipy.optimize import curve_fit
def func(x, a, b, c):
    return a * x**b + c

popt, pcov = curve_fit(func, n, spectral_gap)
print(popt)

# plot the fitted curve
plt.plot(n, func(n, *popt), 'r-')
plt.show()
```
